```python
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv("cleaned_bodyfat.csv")
```

```python
df.describe
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(np.log(df['ABDOMEN']), df['BODYFAT'])
plt.title('Scatter Plot of ABDOMEN vs BODYFAT')
plt.xlabel('Predictor')
plt.ylabel('Target')
plt.show()
```

```python
import seaborn as sns

sns.histplot(df['ABDOMEN'], bins=30, kde=True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='ABDOMEN', y='BODYFAT', data=df)
plt.title('Box Plot of Categorical Feature vs Target')
plt.xlabel('Categorical Feature')
plt.ylabel('Target')
plt.show()
```

```python
sns.pairplot(df[['WEIGHT', 'HEIGHT', 'ABDOMEN', 'BODYFAT']])
plt.show()
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
